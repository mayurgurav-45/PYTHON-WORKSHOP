**Creating Series Object from List**

```python
country = ['India','Pakistan','USA','Nepal','Srilanka']
pd.Series(country)
```

custom index

```python
marks = [67,57,89,100]
subjects = ['Maths','Science','English','Hindi']
pd.Series(marks,index=subjects)
```

setting a name

```python
marks = pd.Series(marks,index=subjects,name='Marks')
marks
```

**Creating Series Object from a Dictionary**

```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks,name='Ajay ke marks')
marks_series
```

```python
marks = {
    'maths':67,
    'english':57,
    'science':np.nan,
    'hindi':100
}

marks_series_2 = pd.Series(marks,name='Ajay ke marks')
marks_series_2
```
